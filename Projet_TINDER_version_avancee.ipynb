## Scope of this project üñºÔ∏è

Data was gathered from participants in experimental speed dating events from 2002-2004. During the events, the attendees would have a four minute "first date" with every other participant of the opposite sex. At the end of their four minutes, participants were asked if they would like to see their date again. They were also asked to rate their date on six attributes: Attractiveness, Sincerity, Intelligence, Fun, Ambition, and Shared Interests.

The dataset also includes questionnaire data gathered from participants at different points in the process. These fields include: demographics, dating habits, self-perception across key attributes, beliefs on what others find valuable in a mate, and lifestyle information. ## 4. Peut-on pr√©voir sa propre valeur ?

#### Quel est l'√©cart entre le nombre de matchs esp√©r√©s et le nombre r√©el de matchs ?

```python
df_comparatif = df_clean.groupby('iid').agg({
    'match_es': 'mean',
    'match': 'sum',
}).reset_index()

df_comparatif['√©cart'] = df_comparatif['match_es'] - df_comparatif['match']

# Trier par √©cart croissant
df_comparatif = df_comparatif.sort_values(by='√©cart', ascending=True)

df_comparatif.head()
```

### Quelle est la moyenne d'√©cart ?

```python
df_comparatif['√©cart'].mean()
```

```python
df_comparatif.value_counts('√©cart')
```

```python
fig = px.histogram(df_comparatif,
                   x='√©cart',
                   title="Distribution des √©carts entre matchs esp√©r√©s et matchs obtenus",
                   labels={'√©cart': '√âcart'},
                   nbins=30)
fig.update_layout(
    yaxis_title="Nombre de participants"
)
fig.show()
```

#### Globalement, apr√®s le premier rdv les participants arrivent globalement √† pr√©dire l'issue. La distribution est Gaussienne, l'√©cart est centr√© autour de 0.

## 5. Pour un second rdv, vaut-il mieux passer au d√©but ou √† la fin ?

```python

df_order = df_clean.pivot_table(index='order', columns='match', aggfunc='size', fill_value=0)

df_order.sort_values(by=1, ascending=False, inplace=True)
df_order
```

```python
# Visualisation du nombre de matchs par ordre de rendez-vous

fig = px.bar(df_order,
             x=df_order.index,
             y=1,
             labels={'x': 'Ordre des rendez-vous', 'y': 'Nombre de matchs'},
             title="Nombre de matchs par ordre de rendez-vous",
             )
fig.show()
```

#### Je constate que l'ordre √† son importance. Les premiers date g√©n√®rent le plus de matchs. Bien que l'ordre ne soit pas pr√©pond√©rant, le nombre de matchs d√©croit avec le temps.

```python


# R√©organiser df_order pour avoir une seule colonne 'match' et une colonne 'count'
df_order_reset = df_order.reset_index()
df_order_long = df_order_reset.melt(id_vars=["order"], value_vars=[0, 1], 
                                    var_name="match", value_name="count")

# Cr√©er l'histogramme avec les barres c√¥te √† c√¥te
fig = px.bar(df_order_long, 
             x="order", 
             y="count", 
             color="match",  # Utiliser la colonne match pour les couleurs
             title="Histogramme de l'Ordre de Passage et du Nombre de Matchs",
             labels={"order": "Ordre de Passage", "count": "Fr√©quence"},
             barmode="group")  # Afficher les barres c√¥te √† c√¥te

# Afficher l'histogramme
fig.show()
```

```python
# Visualisation en heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df_order, annot=True, cmap='Blues', fmt='g', cbar=True)

plt.title('Matrice de Fr√©quences entre Ordre et Match')

plt.show()
```
